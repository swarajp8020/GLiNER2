```python
from gliner2 import GLiNER2

extractor = GLiNER2.from_pretrained("fastino/gliner2-base-v1")
```

 # TASK 1

```python
text = "Rohit Sharma plays for Mumbai Indians in IPL."

result = extractor.extract_entities(
    text,
    ["person", "team", "league"]
)

print(result)
```

```python
result = extractor.classify_text(
    "This phone is amazing but battery is bad",
    {"sentiment": ["positive", "negative", "neutral"]}
)

print(result)
```

# Task 2

```python
text = "iPhone 15 costs ₹80,000 with 128GB storage"

result = extractor.extract_json(
    text,
    {
        "product": [
            "name::str",
            "price::str",
            "storage::str"
        ]
    }
)

print(result)
```

```python
from gliner2 import GLiNER2

extractor = GLiNER2.from_pretrained("fastino/gliner2-base-v1")

text = "Paid ₹750 to Zomato for dinner"

schema = {
    "transaction": [
        "amount::str",
        "merchant::str",
        "category::str"
    ]
}

result = extractor.extract_json(text, schema)
print(result)
```
